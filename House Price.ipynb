```python
def missing_values_table(df):
    mis_val = df.isnull().sum()
    mis_val_percent = 100 * df.isnull().sum() / len(df)
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    mis_val_table_ren_columns = mis_val_table.rename(
        columns={0: 'Missing Values', 1: '% of Total Values'})
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:, 1] != 0].sort_values(
            '% of Total Values', ascending=False).round(1)
    print("Your selected dataframe has " + str(df.shape[1]) +
          " columns.\n" "There are " + str(mis_val_table_ren_columns.shape[0]) +
          " columns that have missing values.")
    return mis_val_table_ren_columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso,ElasticNet
from xgboost import XGBRegressor
from sklearn.metrics import  mean_absolute_error, r2_score,root_mean_squared_error
from sklearn.model_selection import GridSearchCV
df_train['SalePrice_log'] = np.log1p(df_train['SalePrice'])
x_train=df_train.drop(['Id', 'SalePrice', 'SalePrice_log'], axis=1)
y_train=df_train['SalePrice_log']
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.2,random_state=42)
cat_col=x_train.select_dtypes(include=['object']).columns
num_col=x_train.select_dtypes(exclude=['object']).columns
one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)  
scaler=StandardScaler() 
x_train_cat_encoded=one_hot_encoder.fit_transform(x_train[cat_col])
x_test_cat_encoded=one_hot_encoder.transform(x_test[cat_col])
x_train_num_scaled=scaler.fit_transform(x_train[num_col])
x_test_num_scaled=scaler.transform(x_test[num_col])
x_train_scaled = np.concatenate((x_train_cat_encoded, x_train_num_scaled), axis=1)
x_test_scaled = np.concatenate((x_test_cat_encoded, x_test_num_scaled), axis=1)
```

```python
param_grid = {
        'learning_rate': [0.01, 0.05, 0.1],
        'n_estimators': [100,500,600],
        'max_depth': [2, 3],
        'min_child_weight': [1, 3]
    }
model_xgb = XGBRegressor(use_label_encoder=False, eval_metric='rmse', random_state=42,
                               subsample=0.8, colsample_bytree=0.8, gamma=0.1, reg_alpha=0.2, reg_lambda=2)
grid_search = GridSearchCV(model_xgb, param_grid, cv=5, scoring='neg_root_mean_squared_error', n_jobs=-1)
grid_search.fit(x_train_scaled, y_train)
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test_scaled)
final_prediction=np.expm1(y_pred)
actual_prediction=np.expm1(y_test)
rmse = root_mean_squared_error(actual_prediction, final_prediction)
r2 = r2_score(actual_prediction, final_prediction)
mae = mean_absolute_error(actual_prediction, final_prediction)
print("Best Parameters:", best_params)
print("Best Model:", best_model)
print('XGBRegressor')
print("RMSE:", rmse)
print("R2:", r2)
print("MAE:", mae)
print('_'*100)
print('_'*100)
print('LinearRegression')
Lr=LinearRegression()
Lr.fit(x_train_scaled, y_train)
y_pred = Lr.predict(x_test_scaled)
final_prediction=np.expm1(y_pred)    
actual_prediction=np.expm1(y_test)
rmse = root_mean_squared_error(actual_prediction, final_prediction)
r2 = r2_score(actual_prediction, final_prediction)
mae = mean_absolute_error(actual_prediction, final_prediction)
print("RMSE:", rmse)
print("R2:", r2)
print("MAE:", mae)
print('_'*100)
```

```python
print('Ridge')
Ridge_model=Ridge(alpha=1.0)
Ridge_model.fit(x_train_scaled, y_train)
y_pred = Ridge_model.predict(x_test_scaled)
final_prediction=np.expm1(y_pred)
actual_prediction=np.expm1(y_test)
rmse = root_mean_squared_error(actual_prediction, final_prediction)
r2 = r2_score(actual_prediction, final_prediction)
mae = mean_absolute_error(actual_prediction, final_prediction)
print("RMSE:", rmse)    
print("R2:", r2)
print("MAE:", mae)    
print('_'*100)
print('Lasso')
Lasso_model=Lasso(alpha=1.0)
Lasso_model.fit(x_train_scaled, y_train)
y_pred = Lasso_model.predict(x_test_scaled)
final_prediction=np.expm1(y_pred)
actual_prediction=np.expm1(y_test)
rmse = root_mean_squared_error(actual_prediction, final_prediction)
r2 = r2_score(actual_prediction, final_prediction)
mae = mean_absolute_error(actual_prediction, final_prediction)
print("RMSE:", rmse)    
print("R2:", r2)
print("MAE:", mae)
print('_'*100)
print('ElasticNet')
ElasticNet_model=ElasticNet(alpha=1.0, l1_ratio=0.5)
ElasticNet_model.fit(x_train_scaled, y_train)
y_pred = ElasticNet_model.predict(x_test_scaled)
final_prediction=np.expm1(y_pred)
actual_prediction=np.expm1(y_test)
rmse = root_mean_squared_error(actual_prediction, final_prediction)
r2 = r2_score(actual_prediction, final_prediction)
mae = mean_absolute_error(actual_prediction, final_prediction)
print("RMSE:", rmse)    
print("R2:", r2)
print("MAE:", mae)
```

```python
train_pred=Lr.predict(x_train_scaled)
test_pred=Lr.predict(x_test_scaled)
lg_r2_train = r2_score(y_train, train_pred)
lg_r2_test = r2_score(y_test, test_pred)
print('LinearRegression')
print("R2 train:", lg_r2_train)
print("R2 test:", lg_r2_test)
print('_'*100)
print('_'*100)
Ridge_r2_train = r2_score(y_train, Ridge_model.predict(x_train_scaled))
Ridge_r2_test = r2_score(y_test, Ridge_model.predict(x_test_scaled))
print('Ridge')
print("R2 train:", Ridge_r2_train)
print("R2 test:", Ridge_r2_test)
print('_'*100)
Lasso_r2_train = r2_score(y_train, Lasso_model.predict(x_train_scaled))
Lasso_r2_test = r2_score(y_test, Lasso_model.predict(x_test_scaled))
print('Lasso')
print("R2 train:", Lasso_r2_train)
print("R2 test:", Lasso_r2_test)
print('_'*100)
ElasticNet_r2_train = r2_score(y_train, ElasticNet_model.predict(x_train_scaled))
ElasticNet_r2_test = r2_score(y_test, ElasticNet_model.predict(x_test_scaled))
print('ElasticNet')
print("R2 train:", ElasticNet_r2_train)
print("R2 test:", ElasticNet_r2_test)
print('_'*100)
print('XGBRegressor')
xgb_r2_train = r2_score(y_train, best_model.predict(x_train_scaled))
xgb_r2_test = r2_score(y_test, best_model.predict(x_test_scaled))
print("R2 train:", xgb_r2_train)
print("R2 test:", xgb_r2_test)
```
